# Notebook to identify different types of questions

```python
import pandas as pd
import json
```

## Import all questionnaires from github to get an overview of the questions

```python
dfs = []

for i in range(1, 6):
    url = f'https://raw.githubusercontent.com/alexk2206/tds_capstone/refs/heads/main/questionnaires/questionnaire{i}.json'
    df = pd.read_json(url)
    df['options'] = df['options'].apply(lambda x: ', '.join([opt['option'] for opt in x]))
    dfs.append(df)

all_questions_df = pd.concat(dfs, ignore_index=True)
all_questions_df
```

## Count question types to identify the types

```python
type_counts = all_questions_df['type'].value_counts()
print(f'The different kind of questions are:\n{type_counts}')
```

The different types of questions are:

*   SINGLE_SELECT
*   MULTI_SELECT
*   TEXT
*   DATE
*   NUMBER

For every type of question a we want to introduce a text generating function in order to generate possible answers for each question with respect to possible answers
