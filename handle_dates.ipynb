```python
import pandas as pd
from dateutil import parser
from datetime import datetime
import re
```

```python
data = {
    'question': ['When do you want to meet again?', 'ABC?', 'Date_hard'],
    'type': ['DATE', 'TEXT', 'DATE'],
    'options': ['Provide a date', 'TEXT', 'date'],
    'intended_answer': ['2025-06-22', 'TEXT', '2025-06-23'],
    'context': ['We can meet on June 22nd, 2025. That would suit me well', 'TEXT', 'We can meet on June 22nd, 2025. That would suit me well'],
    'predicted_answer': ['June 22nd, 2025', 'HELLO', 'We can meet on June 22nd, 2025. That would suit me well']
}

# Create the DataFrame
df = pd.DataFrame(data)
df
```

```python
def convert_date_format(date_str):
    try:
        parsed_date = parser.parse(date_str)
        return parsed_date.strftime('%Y-%m-%d')
    except Exception as e:
        return date_str

date_regex = r'\b(?:\d{1,2}(?:st|nd|rd|th)?\s+[A-Za-z]+\s+\d{4}|\d{1,2}[./-]\d{1,2}[./-]\d{2,4}|\b[A-Za-z]+\s+\d{1,2}(?:st|nd|rd|th)?,?\s+\d{4})\b'
```

```python
for index, row in df.iterrows():
  if row['type'] == 'DATE':
    match = re.search(date_regex, row['predicted_answer'])
    if match:
      extracted_date = match.group(0)
      formatted_date = convert_date_format(extracted_date)
      df.at[index, 'formatted_date'] = formatted_date
    else:
      df.at[index, 'formatted_date'] = formatted_date
```
